# データ分析に求められること

データ分析は本来の業務の目的のために実施されています．
データ分析の求められることは，その目的に利用できる提案です．
幾ら緻密な分析を行っても，その結果が役に立たなければ「So what」となります．

データ分析の成果物は，統計量やグラフではありません．
それらは，中間成果物です．
データ分析によって凄いことが発見されたとしても，その重要性を伝えられなければ宝の持ち腐れになってしまいます．
分析結果を理解して施策に活用するためには，分析結果についての明確な解釈をする必要があります．

データ分析の最終的な成果物は，分析結果についての考察と提言です．
その提言は，本来業務の方向性に影響を与えます．

## 分析結果の考察

データ分析の結果だけを整理したら，それらについて考察を与えます．
考察のない資料を結果レポートにしてはいけません．
考察をしない分析者は単なるオペレーターです．
いずれ自動化された時には不要な存在になってしまいます．
データ分析の結果に対して，自らの知識と経験を用いて有効な考察を与えることが分析者の付加価値になります．

このフェースでは，分析者の力量が試されます．
単に統計の知識だけでなく，対象の業務や環境についての知識と経験が活かされます．

分析結果についての考察では，次の2点を明確に切り分けて説明します．

- データから分かる事実
- 外部情報による知見
- 推察できること

### データから分かる事実

データから分かる事実は，客観的に言えることだけをまとめます．
そこには，推測を入れてはいけません．

またデータから客観的に言えるとはいえ，真実と言い切ってはいけません．
なぜならば，データは対象となる現象の一つの側面について近似的に観測したものだからです．



### 外部情報による知見

データ分析の多くは業務的用件から発生しています．
そのため，取得したデータ以外の情報が重要な指針になることがあります．

そこで，データ分析対象を含めた環境のついての情報が大変役に立ちます．
業務について深い見識があることもデータ分析を成功させる要因となります．

### 推察できること

そのデータ分析の結果から推察することは分析者の付加価値ですが，根拠の薄い意見も含まれます．
ただし，根拠の薄い意見については，誤解されないように根拠がないことを明記します．

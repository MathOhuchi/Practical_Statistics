# 統計についての今後の学習

この実用統計は，統計学の分野での記述統計を活用しています．
今後の学習のために統計学の分野を体系的に記述します．

統計  
┬─  
├─記述統計  
├─推測統計  
├─ベイズ統計  
└─多変量解析  

## 推測統計

推測統計は，標本集団のデータから母集団の統計量を推測する方法です．
一般的に単に「統計」というと推測統計を指します．

何かを調査するときに，対象全てについてデータを測定することが困難なことが多く，その一部をサンプリングしてデータを取得することが一般的に行われます．
そのとき，サンプリングデータから得られた統計量を母集団の統計量と見做すと，その確からしさが問題になってきます．
推測統計の理論によって，その確からしさを数値化することができます．

推測統計を理解するには，確率および積分についての知識が前提となります．

## ベイズ統計

ベイズ統計は，ベイズ定理の基づいた確率の変化を利用する計算方法です．
理論としては，条件付確率を計算するので，クロス集計の応用と考えられます．

現実的には計算量が膨大になるのでコンピューターが無くては活用が困難です．

ベイズ統計が注目されるようになったのは，機械学習のアルゴリズムにおいてベイズ統計が中心的な役割を果たすためです．
したがって，機械学習，人工知能，ニューラルネットワーク周辺の分野を理解するために必要となります．

## 多変量解析

多変量解析は，複数の変量の間の法則性について分析する手法の総称です．
多変量解析には様々な手法があり，その活用分野の要件に沿うように発展してきました．

代表的な手法には次のものがあります．

- 重回帰分析
- 判別分析
- クラスタリング
- 主成分分析
- 因子分析

その他にも数量化理論や共分散構造分析など多くの理論があります．
全てを網羅的に学習するというよりも，自分の課題を解決するために適切な手法を探して学習する方が合理的です．

また，機械学習においても，内部的に多変量解析の理論を活用している手法も多くあります．

## 統計のための数学

統計分野を深く学習するためには，数学的な基礎を理解している必要があります．
特に次の分野はしっかり学習しましょう．

- 線形代数
- 微分・積分
- 確率

これらの前提としては，集合論や論理学の基礎は必要です．
積分については，測度論およびルベーグ積分が重要になります．

この領域の学習では，「急がば回れ」が鉄則です．

## データサイエンスのキャリア

かつて統計を行っていた人の多くは，統計が専門ではなく，何かの専門分野の補足として統計を行っていました．
例えば，

- 科学研究における観測結果のまとめ
- 生産分野における品質管理
- 都市計画のための人口動態統計
- 販売促進のためのターゲットの絞り込み

等など，多くの実用例があります．
しかし，どの分野においても，主要な業務があって，統計やデータ分析はその補足業務でした．

したがって，キャリアの観点では，まず自分の本来教務のスキルを伸ばすことです．

### データ分析のコンサルティング

業務の分業化および外部リソースの活用の動きにより，データ分析を専門に行うコンサルティング業務が成立するようになりました．
このデータ分析コンサルタントが，データサイエンティストの走りと言って良いでしょう．  
その当時は，統計分析のためのシステムも高価で特別なスキルを備えている必要がありました．

データ分析ツールの発達と普及により誰でもデータ分析に手を付けれるようになり，データ分析業務も本来業務の担当者が担う原点回帰の流れになっています．

### 機械学習のニーズ

人工知能（AI）という言葉が流行り，機械学習のプログラマーがもてはやされています．
既存のプログラミング手法では対応できなかった課題が機械学習によって解決されることもあり，ソフトウェア開発の大きな変革になっています．

機械学習の手法には多変量解析の延長線にある方法も多くあり，統計の知識を有効に活用できます．

その一方で，機械学習の主流はディープラーニングと呼ばれるニューラルネットワーク手法です．
この分野は統計の高度な理論とは一線を画しています．



### データ分析における統計と機械学習の違い

一部の機械学習手法のでは内部的に統計手法を活用しているものも多くあり，統計と機械学習の明確な線引きはできません．
しかし，その違いは精度についての考え方にあります．

統計手法では，データの背景に確率分布があり，正規分布や中心極限定理などから理論的に精度を導きます．

それに対して機械学習では，モデルを作成するときにテスト用のデータを学習用データとは別に用意しておき，テストデータによる検証によって精度を確認します．したがって，論理的というより実践的な精度と考えます．

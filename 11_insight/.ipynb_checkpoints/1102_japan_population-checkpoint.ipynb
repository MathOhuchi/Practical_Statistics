# 分析結果の考察トレーニング

その結果を見て，分析結果の考察について試みます．

## 考察

この人口ピラミッドと代表値から考察として，**データの事実**と**外部情報**や**推測されること**を切り分けて整理します．
これらの考察を混在させないことが，データに対する正しい評価に繋がります．

### データの事実

- 20歳以下の人口構成が年齢が低くなるほど少なくなっているので，少子高齢化が進んでいる．
- ベビーブームによって，最頻値が平均値や中央値と全く異なった値になっている．
- 人口分布や代表値から見ても女性の方が男性より長生きである．

### 外部情報

- 71歳の人口が少ないのは，第二次世界大戦と誕生時期が重なったためである．
- 68歳が最頻値である理由は，第二次世界大戦後の第1次ベビーブームのためである．
- 51歳の人口がすくないのは，丙午の迷信を信じて出産を避けたためである．

### 推測されること

- 少子化傾向が進行すると，現在の社会保障制度の維持が困難になる．
    - 将来の高齢者は，支給年金だけでは生活費を維持できなくなる．
    - 就労者の厚生年金支払額が高くなり，実質的貧困化が進む．
    - 年金制度が破綻し，生活保護老人が増加する．
- 一人暮らし老人が増加する．
    - 健康保険費用が増加する．
    - 健康保険負担金が増加して，実質的貧困化が進む．
    - 自己防衛のために出費を抑えて貯蓄するので，個人消費が落ち込む．
- 労働力確保のため，在日外国人が増加する．
- 労働力を求めて，日本企業が拠点を海外に移設する．
- これらの構造的問題により，日本の経済力が沈下する．

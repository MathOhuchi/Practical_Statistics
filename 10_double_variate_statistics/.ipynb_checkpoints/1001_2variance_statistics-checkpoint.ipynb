# 2変量統計

1変量統計では，その変量の分布を調べることが中心でしたが，2変量統計では，それらの変量の関連性を調べることが中心となります．

<font color='blue'>実用統計において有効な分析手法は2変量統計</font>と言っても過言ではないでしょう．  
変量が沢山ある場合でも，変量の組み合わせを変えて2変量統計を行うことによって，実用に耐えうる知見を得ることが可能です．

2変量統計の先には多変量解析がありますが，多変量解析を実施するよりも適切な2変量統計を実施することを推奨します．
ここで言う「適切な2変量統計」とは，闇雲に全ての変量の組み合わせについて2変量統計を行うのではなく，
2変量統計を行うべき変量を見極めて実施することです．
なぜならば，変量が多い場合，2変量の組み合わせも多くなってしまい，2変量統計の結果資料が膨大になり，
重要な結果が埋没してしまうからです．
ルーティンを熟すのではなく，考えながら進めることが大切です．

2変量統計では，対象となる変量のデータ型によって実施する内容が異なってきます．

| ケース | 変量１ | 変量２ |
|:---:|:---:|:---:|
| 1 | 数値 | 数値 |
| 2 | カテゴリー値 | カテゴリー値 |
| 3 | カテゴリー値 | 数値 |

それぞれのケースに分けて分析方針を確認します．

*****
## 数値×数値

数値×数値の場合，有効な2変量統計は以下の通りです．

- 共分散と相関
- 散布図
- 直線近似

近似直線については，回帰直線が使用されることが多いです．
ただし，回帰直線が適切な近似直線であるとは限らないことに留意する必要があります．

## カテゴリー値×カテゴリー値

カテゴリー値×カテゴリー値の分析は，<font color='blue'>クロス集計</font>が中心になります．

クロス集計表は集計に過ぎませんが，ベイズ統計にも通じる強力な分析手法です．  
クロス集計結果から，様々な条件ごとの比較グラフを作成することができます．

更に，数値データについても階級値によるカテゴリー化によって，クロス集計の分析に持ち込むことができます．

クロス集計として，次のような例があります．

|blood | A | AB | B | O | All |
|:---:|:---:|:---:|:---:|:---:|:---:|
|**Female**| 38 |  3 | 22 | 35 | 103 |
|**Male**  | 33 | 11 | 21 |	26 |  96 |
|**All**   | 71 | 14 | 44 | 61 | 200 |

![カテゴリー値×カテゴリー値](./images/2d_sample2.jpg)

## カテゴリー値×数値

この場合，各カテゴリーごとに数値データを比較することが一般的に行われます．

- 各カテゴリーごとの数値の度数分布
- 各カテゴリーごとの数値のヒストグラム

また，数値の変量を階級化によってカテゴリー値にすることによってクロス集計の分析にすることも可能です．

## 単純で強力な手法

2変量統計で実用上よく利用される手法は，

- 散布図と近似直線
- クロス集計
- グループごとの分布比較

です．

これらを組み合わせてデータ分析を深堀して行く方法は，単純ですが誰にでも理解できる強力な手法です．

ただし，分析結果を読み取って知見を得る洞察力があることが前提で，そのためには対象としている事象や業務についての深い知識が必要となります．

# データのプロファイリング

どのようなデータ分析に於いてもデータを眺める前にそのデータの出生について確認する必要があります．
そのデータが，何時，何処で，誰によって，どのように生成されたか，あるいは，どのように集められたか，という事を明確にします．

分析者がデータの発生源から自分で採取することは，まず期待できません．
誰かにデータを収集してもらっているはずですので，分析する前にデータの発生・収集に関わる特徴を確認して文書化します．

## データの種類

コンピュータの発達により，データの種類も豊富になりました．

- 数値
- 文字列
- 画像
- 音声
- 動画
- 複合構造データ

この講座では，このうち数値データと文字列データのみについて扱います．

## データの不完全性

分析者は手元にあるデータが分析したい状況を完全に捕捉していると考えてはいけません．
データは，観測対象のある側面しか捉えていません．
したがって，データを緻密に分析したからと言って観測対象を正確に表現できたと過信してはいけません．
データ分析の心構えとしては，常にデータを疑うことです．
そこで，データに刷り込まれた歪みを確認するために幾つかの事象について事前に確認します．

- データの発生目的
- 母集団は何か
- 紙データと電子データ
- システムの特徴

### データの発生目的

まず明確にすることは，そのデータの発生目的を知ることです．

- 今回のデータ分析のために集められたデータ
- 他の目的のために集められたデータを流用する

今回のデータ分析のために集められたデータの例としては，アンケート調査の結果データです．
アンケート分析では，アンケートの設計と調査対象者の選定が重要になってきますが，最初から今回の目的のデータ分析のために実施されます．

他の目的のために集められたデータの例としては，スーパーマーケットのPOSデータです．
分析例としては，POSデータから購買パターンを調べるバスケット分析などがあります．
このPOSデータは，店舗の業務処理のためのデータであり，データ分析を主眼としたデータではありません．
たまたま，そのデータを活用すればバスケット分析ができるので流用したということです．

### 母集団は何か

手元にあるデータが，調査対象の全てについてのデータであるのか，その一部の対象についてのデータであるかによって，分析方針が異なります．
ここで調査対象全体を母集団と言い，その一部は標本集団あるいは単に標本と言います．
標本は母集団からバランスよく選ばれていることが理想です．
しかし，一般的には標本に偏りがあります．この偏りをバイアスと言います．


### 紙データと電子データ

データと言うと電子化されている前提で話をしますが，最初にデータを記述する媒体が紙という場合があります．
オリジナルが紙の場合，そのデータを電子化する必要があります．
電子化プロセスにおけるご記入などのトラブルには注意深く対応しなければなりません．
電子化するときに，紙上で何も書かれていない項目をどのようにするかというデフォルト値の対応についても神経を使わねばなりません．


### システムの特徴

最初から電子化されているデータについては，そのデータがどのようなシステムのものなのかを確認します．
特に，POSデータのようにデータ分析を目的としていないシステムから流用した場合は，分析のためのデータが揃っていない可能性もあります．
したがって，データの抽出元のシステムがどのような特性のものなのかを確認して，その影響も考慮します．

# データのプロファイリング

どのようなデータ分析に於いてもデータを眺める前にそのデータの生い立ちについて確認する必要があります．
そのデータが，何時，何処で，誰によって，何のために，どのように生成されたか，あるいは，どのように集められたか，という事を明確にします．

分析者がデータの発生源から自分で採取することは，まず期待できません．
誰かにデータを収集してもらっているはずですので，分析する前にデータの発生・収集に関わる特徴を確認して文書化します．

## データの不完全性

分析者は手元にあるデータが分析したい状況を完全に捉えていると考えてはいけません．
データは，観測対象のある側面しか捉えていません．
したがって，データを緻密に分析したからといって観測対象を正確に表現できたと過信してはいけません．
データ分析の心構えとしては，常にデータを疑うことです．
そこで，データに刷り込まれた歪みを確認するために幾つかのことを事前に確認します．

- データの発生目的
- 母集団と標本
- 紙データと電子データ
- システムの特徴

### データの発生目的

まず明確にすることは，そのデータの発生目的です．

- 今回のデータ分析のために集められたデータなのか，
- それとも，他の目的のために集められたデータを流用する

今回のデータ分析のために集められたデータの例としては，アンケート調査の結果データです．
アンケート分析では，アンケート設計と調査対象者の選定が重要になってきますが，最初から今回の目的のデータ分析のために実施されます．
この場合は，アンケート設計からデータ収集が大きな作業になります．

他の目的のために集められたデータの例としては，スーパーマーケットのPOSデータです．
分析例としては，POSデータから購買パターンを調べるバスケット分析などがあります．
このPOSデータは，店舗の業務処理のためのデータであり，データ分析を主眼としたデータではありません．
たまたま，そのデータを活用すればバスケット分析などのマーケッティング活用ができるので流用したということです．

### 母集団と標本

手元にあるデータが，調査対象の全てについてのデータであるのか，対象の一部のデータであるかによって，分析方針が異なります．
ここで調査対象全体を<font color='blue'>母集団</font>と言い，対象の一部は<font color='blue'>標本</font>あるいはサンプルと言います．

標本は母集団からバランスよく選ばれていることが理想です．
バランスよくデータを抽出することを<font color='blue'>無作為抽出</font>と言います．
しかし，一般的には標本に偏りがあります．この偏りを<font color='blue'>バイアス</font>と言います．

### 紙データと電子データ

データと言うと電子化されている前提で話をしますが，最初にデータを記述する媒体が紙という場合があります．
オリジナルが紙の場合，そのデータを電子化する必要があります．

また，電子化プロセスにおける誤記入などのトラブルには注意深く対応しなければなりません．
電子化するときに，紙上で何も書かれていない項目をどのようにするかという<font color='blue'>欠損値</font>の対応についても神経を使わねばなりません．

### システムの特徴

最初から電子化されているデータについては，そのデータがどのようなシステムのものなのかを確認します．
特に，POSデータのようにデータ分析を目的としていないシステムから流用した場合は，分析のためのデータが揃っていない可能性もあります．
したがって，データの抽出元のシステムがどのような特性のものなのかを確認して，その影響も考慮します．

実際にある例について紹介します．小売のPOSデータから売れ筋を分析するのですが，POSの品目として総菜A，総菜Bなどがあります．この総菜のラベルと価格は変化ないのですが中身が食材の調達によって全く変わってしまいます．具体的な中身の情報はPOSデータには含まれていないので，これだけでは肝心なことが分かりません．

## データの種類

コンピュータの発達により，データの種類も豊富になりました．

- 数値
- 文字列
- 画像
- 音声
- 動画
- 複合構造データ

この講座では，このうち数値データと文字列データのみについて扱います．

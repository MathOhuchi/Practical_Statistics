# プロジェクト意識

近年の傾向として，何事においてもデータ分析を行うことが推奨されますが，安易なデータ分析が弊害になることもあります．
ビッグデータという言葉が流行っているように，私たちは多くのデータに囲まれています．
そして，そのデータを使わないことは宝の持ち腐れと言われてしまいます．

またデータ分析ツールについても，統計専門家でなくても操作方法についての適度なトレーニングを行えば，
単純な集計から高度な分析まで実施することが可能になっています．

したがって，何か課題が発生すると，意思決定者は部下に対して安易にデータ分析を指示するケースが増えています．

そこには，データ分析を正当化する3つの事象があります．

- 大量のデータが利用可能になった
- 分析ツールが利用しやすくなった
- 客観的説明が要求される

データ分析には多くのメリットがあります．
しかし，使い方を間違えると単なる時間と労働力の消費だになるけではなく，間違った方向にガイドしてしまう危険性もあります．

## 本来の目的

データ分析の事案が発生する理由は何らかの業務課題にあります．
その業務課題とは，単に企業の置ける業務課題だけでなく，広く人間の社会活動に関する課題と考えることができます．

業務課題が一旦提起されると，業務分析を行います．
業務分析の結果として，様々な領域に特定された課題にブレークダウンされます．
そして，そのブレークダウンされた課題の一つとしてデータ分析課題が定義されます．

したがって，データ分析の結果の評価は，本来の業務課題における評価に依存します．
さらに，結果の評価だけでなく，分析の方針や方法についても，本来の業務課題に依存して決まってきます．

この本来の目的を無視してデータ分析を推進することは避けなければなりません．

![本来の目的とデータ分析](./images/purpose.jpg)

### 業務目的に沿ったデータ分析

上図にあるように元々の業務課題に対して業務分析を行った結果，様々な分野における課題への落とし込まれます．
そこで考えるべきことは，どの分野の課題を優先的に解決することが効果的なのかということです．

データ分析を行っても効果が少ないならば，データ分析を断念すべきかもしれません．
肝心なことは，データ分析結果をどのように施策に反映するかです．
施策に反映できなければ，データ分析の実施意味がありません．

![施策への反映](./images/action_chain.jpg)

### 需要予測の例

ある季節性商品は市場の状況によって出荷量がかなり左右されます．
そして，その製品の生産ラインは2ヶ月前に生産計画を立てます．
この生産計画が売上量を超えると不良在庫を抱えることになります．
逆に生産計画を上回る需要があっても品切れとなり，販売機会の損失となります．

そこで，その企業は精密な生産計画を立案するために需要予測のデータ分析を外部コンサルタントに委託しました．
委託されたコンサルタントは当然ながら最適となるように需要予測を実施して，結果を納入します．
その結果を活用すると，それまでの生産計画よりも正確性が向上しました．



機械学習によって，次のような需要予測モデルを構築しました．

> <font color=green>売上予測 $= f($ 過去の売上実績, 経済状況, 社会状況, 等など $)$ </font>

このような予測モデルを構築するには，機械学習のスキルだけでなく，変数となるデータ選択において商品を取り巻くビジネス構造などに精通していることが必要となります．

#### 局所最適化の罠

この需要予測は，それ自体は一定の効果があったので良かったのですが，この施策に拘ることがより良い解決策を逃すことにもなります．
依頼を受注したコンサルタントは，需要予測を目的として受注しているので，需要予測の最適化に集中します．
問題は，発注側の業務分析でデータ分析以外の対応策についての検討が無く，安易に需要予測に頼った点にあります．


#### パラダイムシフト

実は，この企業の場合，サプライチェーンの大きな課題があり，そのために2ヶ月前に生産計画を確定する必要がありました．
そして，そのことが問題の本質であることに気付き，サプライチェーンの改革に取り組みました．
その結果，生産計画の確定が2ヶ月前から2週間前へと劇的に短縮することに成功しました．

すなわち，需要予測に頼るという重たい体質から，直近の推移に基づいて生産計画を柔軟に修正するSense & Responseという体質へのパラダイムシフトを果たしました．


![パラダイムシフト](./images/paradigm_shift.jpg)

このサプライチェーン改革は，需要予測の精度も向上させました．
2か月先の予測と2週間先の予測の精度を比べたら，当然の結果です．
さらに良いことは，突発的な状況によって出荷見込みが変動した場合のインパクトも少なくなります．

この例からも，需要予測に拘らず広範囲に課題解決策を考える事が重要であることが分かります．


## データ分析の客観性

データ分析には様々な人の思惑が絡みます．
それらの人達の願望をデータ分析結果に反映してしまう可能性があります．
しかし，本来のデータ分析はデータから導かれる事象であり，願望が入ってはいけません．
むしろ，客観的事象と願望とのギャップを明確にして，そのギャップに対する対応を検討することが健全なプロセスと言えます．

そのような例を見てみましょう．

### 結論ありきの分析

これは架空の話ですが，ある地域において自動車道の建設計画があります．
そこの自動車道を作ることは，生活者の利便性も上がり，なにより緊急時の人命救助や防災に役立ちます．
しかし，高額な建設費が発生するので地元自治体の財政負担となります．

そこで，自動車道路を建設した後の20年間の交通量の推定を行い，地元の産業振興を含めた採算性のアセスを行いました．
その結果で採算性が認められたら，自動車道の建設計画を実施することにとしました．

その採算性アセスは某外郭団体に委託され，そこで専門的な数理モデルを用いて計算します．
その外郭団体は，行政から委託を受けて建設効果アセスを行っている専門能力のあるところです．
しかし，分析を委託側は自動車道路建設の推進派であり，自動車道路建設を実現するためのエビデンスがほしかったのです．


ここでの問題点として，データ分析が結論ありきで進められていることです．
つまり，自動車道路を建設した場合に採算性が認められるような結論を導くことが目的になっていることです．
この外郭団体が建設に否定的な結果を提出すると今後同じような依頼が来なくなることを懸念して，建設推進に都合の良い結論を導いたとしたらどうでしょうか．
結局，正統な分析結果でなくても，そのことを指摘できる第三者としての専門家がいなければデータ分析の悪用でしかありません．

ではどうすべきかというと，当然ですがデータ分析では客観的に確からしい採算性のアセス結果を出すべきです．
その分析結果で採算性が認められないという結論が得られたとしたら，その前提の下で，それでも地元の防災や生活での便利性とを考慮した総合的な判断を行うべきです．

### 予言効果

自部門の売上などを予測するデータ分析に関しては，予測結果を自部門の営業員が知ることによって，実際の売上結果が予測値の近くに落ち着く可能性が高くなることが知られています．
これを予言効果といいます．
営業員の心理として，予測値を下回ると営業努力が足りないと思われることを危惧して頑張ります．
逆に予測値を超えることが分かると満足して気を緩めてしまいます．

この予測効果を利用して，予測値を高めに出すように指示する部門長もいるかもしれません．
しかし，正しく理解して判断するという基本からすると，この対応は間違っています．
データ分析を理解している人が見たら，改竄にあたるからです．
データ分析結果を歪めたら信用を失ってしまいます．
そして，一度失った信用の代償は大変大きなものになるでしょう．

どうすべきかというと，これも当然ですが，予測値は正しく求める．
そして目標値を設定して，これは予測値より高い値になるでしょうが，そこに何故ギャップがあり高い目標値を目指さなければならないのかをしっかり説明することです．
部門長が説明責任を果たすことが強い組織の育成になります．

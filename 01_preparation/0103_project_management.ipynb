# データ分析プロジェクトの全体

データ分析プロジェクトにおいて注目されるフェーズは統計的手法を用いたデータ分析の中核部分でしょうが，プロジェクトの観点からすれば高度な分析を行うフェーズは少ししかありません．
データ分析をプロジェクトとして管理することが大前提になることを認識して計画，実施，結果活用の提言を行います．

## データ分析プロジェクトのフェーズ

本来の業務目的に沿った業務分析の結果を受けて，データ分析プロジェクトは明確な目的を持って開始されます．

データ分析プロジェクトの大まかなフェーズは次のようになります．

- 計画
- データ収集
- データクレンジング
- 本格的なデータ分析
- 結果解釈～活用への提言

これらの標準的な作業量はつぎのような構成です．
ただし，データ収集が非常に膨大な作業になることもあります．

![プロジェクトのフェーズ](./images/project_phase.jpg)

ここでのポイントは，統計を駆使した本格的なデータ分析を行う前にデータクレンジングという処理が必要になることです．
すなわち，データ分析にはITスキルが必須だということです．
私たちは，統計理論を学ぶと同時にデータ処理のためのプログラミングもマスターすることが求められます．

## スキルと知識

データ分析プロジェクトに必要なスキルと知識は多くあります．
統計だけを学習すれば良いというわけではないことが分かります．

- 本来業務の知識
- 統計の知識
- プログラミング・スキル
- データ分析ツールの知識
- プロジェクト管理

### 本来業務の知識

データ分析が必要となった業務についての知識，あるいは，ビジネスならばその業界の知識は必要不可欠なものです．
最終的なデータ分析の洞察力の決め手になります．

### 統計の知識

データ分析ツールの使い方を覚えれば，統計解析を実施することは可能です．
しかし，統計知識が無ければ，重要な判断を行う確信を持つことができません．
結果を解釈するためにも統計の基礎知識は必要です．

### プログラミング・スキル

データ分析に使用するデータ加工の加工やクレンジングにはプログラムを利用します．
私たちはPythonを採用しますが，一般的に普及しているプログラム言語ならば問題は無いでしょう．
私たちがPythonを採用した理由は，データ分析ではPythonが最も利用されているからです．

### データ分析ツールの知識

データ分析ツールには様々なものがあります．
有名なSAS，SPSS，R言語などツールには様々なものがありますが，
ここでは，Pythonのライブラリーをデータ分析ツールとして考えています．
ツールについてはマニュアルを参照しながら利用できるレベルで大丈夫です．

### プロジェクト管理

プロジェクト管理では次の要素が重要になります．

- スケジュール管理
- コスト管理
- 品質管理
- 期待値管理
- リスク管理

これらについては，通常のプロジェクトと同じと考えてください．
データ分析プロジェクトの場合，分析結果によって成果の価値が大きく変わります．
緻密な分析結果であっても，「so what」にならないことが肝心です．

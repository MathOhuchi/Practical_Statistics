# データ分析の種類

データ分析プロジェクトを大きく分類すると次の2通りになります．

- <font color=blue>求解</font>：求める結果が明確に定義されていて，その答えを求める分析
- <font color=blue>探索</font>：データの中身を探りながら何らかの課題や問題点を見つけ出す調査


*****
## 求解データ分析

データ分析を行う前提として，最終的に取得するものが決まっている場合を<font color=blue>求解</font>と言います．
具体的な例を幾つか記載します．

- パラメータの最適化
- 代表値の推定と検定
- ランキングと抽出

これは，データ分析で得られた結果が次の施策で活用されることが明確になっています．
この<font color=green>結果が活用される</font>ことがデータ分析の意義を明確にし，価値を高めています．

### パラメータの最適化（例）

ある植物の生育の良し悪しが複数の環境要因によって決まります．それぞれの要因を変量として，幾つかのパターンの生育実験を行いました．全てのパターンの実験を行うことは不可能なので，幾つかの条件で実験した結果を測定しました．
この測定結果を元にして，その植物の育成に最適な変量のパターンを推定します．

このデータ分析は，変量の組み合わせの最適値を求める求解です．

この結果として求まった変量の最適パターンは，その植物の繁殖プラントで利用されます．施策において具体的に使用されます．

### 代表値の推定と検定（例）

ある生産ラインで製造される製品について，ロットごとに99.99%の精度で品質基準を満たしていればそのロットは出荷となります．
品質の確認はサンプリング検査によって行われます．
サンプリング検査の結果からロット全体が99.99%以上の確率で品質基準を満たしていることを検定します．

このデータ分析で求まることは，その標本集合の品質値の分布です．この分布からロット全体の品質値の分布を求めて判定を行います．

この結果によって，ロット全体の出荷が決まります．品質基準に満たない場合は原因調査および生産ラインの点検が行われます．データ分析結果が確実に利用されています．

### ランキングと抽出（例）

登録された顧客の情報および膨大な購買履歴があります．
このデータを活用して，発売予定の新製品についての購買確率を全ての顧客について予測します．顧客は購買確率の高い順にランキングされて，上位の顧客層を抽出することができます．

この結果は新製品の販売促進のダイレクトマーケティングの対象者として利用されます．

## 探索データ分析

ある状況について，そこに潜む問題点を発見したり，新しい可能性を発見するためのデータ分析を<font color=blue>探索</font>と言います．
これは，データを使った調査とも言えます．
このように何かを発見するという行為の背景には，<font color=green>通常ということをよく理解している</font>ことが前提になっています．
この通常への理解なしに，探索を行うことはできません．

探索データ分析には次のような目的があります．
- 万が一の危機を早期に発見する．
- 新しいニーズを発見する．

探索は何かを発見するためのデータ分析なので，何も発見できないこともあります．
また何かを発見できたとしても，その結果をどのように施策へ反映するかという課題が残ります．

- 何も発見できない場合，どこで分析を止めるか．
- 発見した事象をどのように施策に反映させるか．

データがあるからと言って，闇雲に探索を行うことは，時間と労力の無駄遣いです．
探索データ分析をどの程度実施するかは，その組織の置かれている状況によって決まってきます．

## 計画を立てる

データ分析を行うときに，これも一つのプロジェクトあるいはタスクとして認識しましょう．
計画が明確でない場合，データ分析作業がズルズルと進んでしまいます．
とくに探索では歯止めを最初に明確にすることが肝心です．

データ分析の計画として，少なくとも次の事を明確にして，文書化しておきましょう．

- 本来の業務の目的
- データ分析のゴール
- 作業期間と作業工数
- 成果基準

この計画文書は大袈裟なものでなく，ほんの数行のメモであっても構いません．
この業務課題に関わる人と共有して，認識を一致させておきましょう．
